```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
wine= pd.read_csv("wine.csv")
```

```python
wine1=wine.drop(['Type'], axis=1)
wine1.keys()
```

```python
wine1.describe()
```

```python
sns.pairplot(wine.iloc[:,0:12])
```

```python
from sklearn.preprocessing import StandardScaler
std_df = StandardScaler().fit_transform(wine1)      
std_df.shape
```

```python
from sklearn.decomposition import PCA
pca_std = PCA(random_state=14, n_components=3)
pca_std_df= pca_std.fit_transform(std_df)
print(pca_std_df)
```

```python
pca_std_df.shape
```

```python
print(pca_std.singular_values_)
```

```python
print(pca_std.explained_variance_ratio_*100)
```

```python
cum_variance = np.cumsum(pca_std.explained_variance_ratio_*100)
cum_variance
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 8))
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_) 
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method - K-means with PCA Clustering')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df_norm)

y_kmeans
```

```python
pca_kmeans =pd.concat([wine1.reset_index(drop = True), pd.DataFrame(df_norm)], axis= 1)
pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']
pca_kmeans['Segment Kmeans PCA'] = y_kmeans
print(pca_kmeans)
```

```python
pca_kmeans['Segment'] = pca_kmeans['Segment Kmeans PCA'].map({0:'First', 1:'Second', 2:'Third'})
```

```python
x_axis = pca_kmeans['Component 2']
y_axis = pca_kmeans['Component 1']
plt.figure(figsize= (10,8))
sns.scatterplot(x_axis, y_axis, hue=pca_kmeans['Segment'],palette=['g', 'r','c'] )
plt.title('Clusters by PCA Componets')
plt.show()
```

```python
x_axis = pca_kmeans['Component 2']
y_axis = pca_kmeans['Component 3']
plt.figure(figsize= (10,8))
sns.scatterplot(x_axis, y_axis, hue=pca_kmeans['Segment'],palette=['g', 'r','c'] )
plt.title('Clusters by PCA Componets')
plt.show()
```
